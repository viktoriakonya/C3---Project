# Analysis of board game ratings

### Business question: What characteristics of the board games determine their ratings? 


### Introduction
In this analysis we are going to investigate the important predictors of the ratings of the board games. Our main question is that which characteristics of the board games impact their ratings and what is the direction of the relationship. Our outcome variable is the **(average) ratings** and we considered the following predictors:

* Maximum playtime
* Minimum playing age
* Minimum and maximum number of players
* Number of users rated the board games
* Average learning and strategy complexity score
* Category of the board games (category flags were created for the most frequent categories as explained below)
* Year when the board games were published

We will start our analysis with an explanatory analysis of the outcome variable and the predictors in order to explore our dataset. First, we will explore each variable individually in order to spot data issues and implement the necessary transformations. Then we will explore the relationship of each predictor with the rating to better understand what patterns we can expect. Finally, we will examine the relationship in regression framework and summarize our results.

### What the code contains:
* Univariate analysis of the outcome variable and each predictor (including data cleaning, feature engineering, variable recodings and summary statistics).
* Bivariate analysis of the relationship between the predictors and the outcome.
* Regression analysis

#### 3.8. Category

### The category of the board games is defined as a list of categories. For further analysis we need to define dummies for each category and select the most frequent types.
For this we need to:
* Omit those records where the category is missing (there were 45 records with missing category)
* Then we need to create a corpus with the unique categories from all board games.  
* Finally, we need to transpose the created dataset and select the most frequent categories (there were about 83 unique categories so we decided to keep only the top 20 most frequent categories) and create dummies.
* Finally we can join back this dataset to the original data.

```python
# Check board games categories
board_games['category_base'].head()
```

**3.8. a) The first step is to remove the records with missing categories and create a new temporary dataset.**

```python
# Remove nulls
board_games_temp = board_games[pd.notnull(board_games['category_base']) == True]
print(len(board_games))
print(len(board_games_temp))
board_games_temp.head()
```

**3.8. b) Then we need to clean the categories. We will concatenate the category names (eg. 'Card game' will become 'cardgame') and remove the commas so that the categories will be separated by a space.**

```python
# Remove commas and split categories by word 
def clean_text(text):
    # Lowercase
    text = text.lower()
    # Remove spaces and commas
    text = re.sub(' ','',text) 
    text = text.replace("'", "")
    text = re.sub(',',' ',text)
    return text

board_games_temp['category_base_clean']= board_games_temp['category_base'].apply(clean_text)
board_games_temp['category_base_clean'].head()
```

**3.8. c) Next, we create the word corpus with the unique categories. Our dataset has 83 unique categories. Some samples are shown below.**

```python
# Create word corpus
corpus = []
for ind in board_games_temp.index:
    for word in board_games_temp['category_base_clean'][ind].split():
        if word not in corpus:
            corpus.append(word)
            
print(corpus[0:10])
print(len(corpus))
```

**3.8. d) Next, we create a dictionary where we count the number of occurances of the categories ( it is one if a particular category appears in the category list of the board game). We also convert it to dataframe.**

```python
# Create dictionary with words by id
my_dict = dict.fromkeys(board_games_temp['game_id_base'])

for ind in board_games_temp.index:
    
    name = board_games_temp['game_id_base'][ind]
    col_appended = []
    
    for word in corpus: 
        
        order = len(word)
        words = "".join([word for word in board_games_temp['category_base_clean'][ind].split() if len(word) == order])
        
        col = len(re.findall(word, words))
        col_appended.append(col)
        
    my_dict[name] = col_appended

    # Create dataframe
dummies_pre = pd.DataFrame(my_dict)
dummies_pre.head()
```

**3.8. e) Last we need to transpose the matrix and crete the final dataset with the dummies.**

```python
# Transpose df
dummies_transposed = np.transpose(dummies_pre)

# Assign filnames as column names
dummies_transposed.columns = corpus

# Add index  to columns
dummies_transposed.reset_index(inplace=True)
dummies_transposed = dummies_transposed.rename(columns = {'index':'game_id_base'})
dummies_transposed.columns += '_d'

pd.set_option("display.max_columns", None)
dummies_transposed.head()
```

**3.8. f) Now we can join it back to the main dataset.**

```python
# Join datasets
board_games = pd.merge(left = board_games, 
      right = dummies_transposed, 
      how = "left",
      left_on = "game_id_base",
      right_on = "game_id_base_d")
board_games.head()
```

**3.8. g) Let's check the most frequent categories.**

```python
dummies = board_games[board_games.columns[pd.Series(board_games.columns).str.endswith('_d')]].sum().sort_values(ascending = False).head(21)
dummies = pd.DataFrame(dummies)
dummies.rename(columns={'0': 'count'})
dummies.reset_index(inplace=True)
dummies = dummies.rename(columns = {'index':'dummy', 0: 'count'})
dummies = dummies[dummies['dummy'] != 'game_id_base_d']
# dummies
```

```python
fig = sns.barplot(y='dummy', x='count', data=dummies, palette = "mako")
plt.xlabel("Count")
plt.ylabel("Dummies")
plt.title("Frequency of most common categories") 
plt.show(fig)
```

**3.8. i) We will remove those category dummies that are not among the most frequent dummies.**

```python
# Get all dummies minus the most frequents
all_dummies = board_games[board_games.columns[pd.Series(board_games.columns).str.endswith('_d')]].columns
keep_dummies = dummies['dummy'].unique()
remove_dummies = list(set(all_dummies) - set(keep_dummies))
# remove_dummies
```

### 4.1. Continous predictors

In case of the continous predictors we examined the scatterplot and the correlation matrix of the outcome and each predictor.

The main findings are the following:
* The ratings are in positive relationship with the maximum playtime: the higher the playtime is the higher rating we can expect.
* The ratings decrease with the maximum number of players: the higher the number of the maximum number of players required for the game the lower ratings we can expect. Somewhat similar pattern can be observed in case of the minimum number of players.
* Regarding the level of strategy and learning complexity, we can again see an increasing pattern.
* Finally, board games with more user ratings are expected to have higher ratings.

```python
predictors = board_games_short.loc[:, board_games_short.columns.isin([
    'max_playtime_base', 
    'max_players_base',
    'min_players_base',
    'users_rated_log_base',
    'average_learning_complexity_api',
    'average_strategy_complexity_api',
    ])].columns.values.tolist()
predictors
```

```python
fig, axes = plt.subplots(nrows=len(predictors)//2,ncols=2, figsize=(20,20))
i = 0
for x in predictors:
    sns.regplot(data=board_games_short, x=x,y='average_rating_base',ax=axes[i//2, i%2])
    i += 1
plt.show()
```

From the correlation matrix we can see that the number of users who rated the board games are in the strongest correlation with the ratings. ## 5. Regression analysis

For the regression analysis we need to convert the categorical variables into dummies.

```python
# Create dummies
board_games_short_year_dummies = pd.get_dummies(board_games_short['year_published_category_base'])
board_games_short_year_dummies.head()
```

```python
# Create dummies
board_games_short_age_dummies = pd.get_dummies(board_games_short['min_age_category_base'])
board_games_short_age_dummies.head()
```

```python
board_games_short = pd.concat([board_games_short, board_games_short_year_dummies, board_games_short_age_dummies], axis=1)
board_games_short.head()
```

First, let's define the final predictor list which includes all the dummies created above (we excluded the reference groups from the dummies).

```python

# Define predictor list
x1 = board_games_short[[
    'max_players_base',                 
    'max_playtime_base',                              
    'min_players_base',
    'average_learning_complexity_api',
     'users_rated_log_base']]


x2 = board_games_short[[
    'max_players_base',                 
    'max_playtime_base',                              
    'min_players_base',
    'average_learning_complexity_api',
    'users_rated_log_base',   
    # 'average_strategy_complexity_api', # collinearity
    # '1951-1960', # Reference
    '1961-1970',
    '1971-1980',
    '1981-1990',
    '1991-2000',
    '2001-2010',
    '2011-2020',
    # '0-7', # Reference
     '8-12',
     '13-17',
     '18+']]

x3 = board_games_short[[
    'max_players_base',                 
    'max_playtime_base',                              
    'min_players_base',
    'average_learning_complexity_api',
    # 'average_strategy_complexity_api', # collinearity
    # '1951-1960', # Reference
    '1961-1970',
    '1971-1980',
    '1981-1990',
    '1991-2000',
    '2001-2010',
    '2011-2020',
    # '0-7', # Reference
     '8-12',
     '13-17',
     '18+',
     'users_rated_log_base',          
     'economic_d',
     'negotiation_d',
     'cardgame_d',
     'fantasy_d',
     'abstractstrategy_d',
     'medieval_d',
     'ancient_d',
     'exploration_d',
     'bluffing_d',
     'sciencefiction_d',
     'dice_d',
     'fighting_d',
     'adventure_d',
     'wargame_d',
     'humor_d',
     'partygame_d',
     'animals_d',
     'movies/tv/radiotheme_d']]

y = board_games_short[['average_rating_base']]
```

We will run 3 OLS models. We start with a regression model that contains only the continous predictors. Then in the next model we add the categorical variables to the model. Finally, we add the category dummies to the regression and interpret the conclusions.

In order to interpret the results we will create nice summary tables of regression output with the **statmodels** package.

#### 5.1. Regression model with the base set of predictors

```python
import statsmodels.api as sm

x_const = sm.add_constant(x1) # Add constant
est = sm.OLS(y, x_const)
est2 = est.fit()
print(est2.summary())
```

The R-squared is about 24% of the simplest model and all predictos are statistically significant. As expected from the two-variable analysis the number of players (both the minimum and maximum) are in inverse realtionship with the rating and the rest of the predictors are associated with higher ratings. 

#### 5.2. Extended regression model with the categorical predictors

```python
x_const = sm.add_constant(x2) # Add constant
est = sm.OLS(y, x_const)
est2 = est.fit()
print(est2.summary())
```

As we added more predictors to the model the explonatory power increased to 41%. Again, as we expected, newer games are associated with higher ratings compared to games launched in the 50's. If we look at the age groups where our reference group is the games without minimum age requirement (or suitable for very small kids) we can see that higher rating can be expected for games where the minimum age requirement is higher. Regarding the games for adult audience it seems that the pattern does not hold. 

#### 5.3. Extended regression model with the categorical predictors and category dummies

```python
x_const = sm.add_constant(x3) # Add constant
est = sm.OLS(y, x_const)
est2 = est.fit()
print(est2.summary())
```

As we added the 20 category dummies to the model, the explanatory power only increased by about 4 percentage point. The parameter estimates are also impacted by the additional predictors: the coefficient of the minimum required players are not significant in the last model and the magnitude of the rest of the predictors from the previous model slightly decreased. From the category dummies we can see that boardgames with categories such as abstract strategy, fighting and wargame have on avarage higher ratings than board games without these categories while card game, dice, humor and board games related to movies seem to be less popular.

Note that wargames especially may be affected by discarding extreme values earlier, these tend to be very long games. However, upon runing the same regression without discarding these values the results significantly didn't change.

### 6. Conclusions

In this analysis, we examined the relationship between the average rating and the characteristics of the board games. We found that the number of users who rated the board games and the learning (and strategy) complexity score of the games are in the strongest relationship with their ratings. We also saw that newer games tend to have higher scores and and the longer the playtime of the games the higher ratings we can expect. Regarding the minimum age requirement of the board games we can say that board games with a minimum requirement of at least 8 years are expected to have higer ratings than board games without minimumum age reqirements.
